## Section 4 — Document Loading & Preprocessing

````python
@dataclass
class SBPPage:
    doc_name    : str
    page_number : int
    text        : str
    category    : str
    short_name  : str
    section     : str = ""
    char_count  : int = 0

    def __post_init__(self):
        self.char_count = len(self.text)
````

````python
def clean_pdf_text(text: str) -> str:
    if not text:
        return ""
    lines   = text.split("\n")
    cleaned = []
    seen_short = {}

    for line in lines:
        s = line.strip()
        if not s:
            continue
        if re.match(r'^[\d\s|]+P\s*a\s*g\s*e\s*$', s):
            continue
        if re.match(r'^Page\s+\d+\s+of\s+\d+', s, re.IGNORECASE):
            continue
        if len(s) < 60:
            key = s.lower()
            seen_short[key] = seen_short.get(key, 0) + 1
            if seen_short[key] > 2:
                continue
        cleaned.append(s)

    text = " ".join(cleaned)
    text = re.sub(r'\s{2,}', ' ', text)
    text = re.sub(r'(\w+)-\s+(\w+)', r'\1\2', text)
    return text.strip()
````

````python
def detect_section(text: str) -> str:
    patterns = [
        r'(REGULATION\s*[\u2013\-]\s*\d+[^\n]{0,60})',
        r'(Regulation\s*[\u2013\-]\s*\d+[^\n]{0,60})',
        r'(Regulation\s+[A-Z]\-\d+[^\n]{0,60})',
        r'(Section\s+[IVXLC]+[^\n]{0,60})',
        r'(Part\s+[A-Z][^\n]{0,40})',
        r'([A-Z][A-Z\s/]{5,40}:)',
    ]
    for p in patterns:
        m = re.search(p, text)
        if m:
            return m.group(1).strip()[:100]
    return "General"
````

````python
def load_sbp_documents(config: dict) -> List[SBPPage]:
    sbp_dir  = Path(config["SBP_DOCS_DIR"])
    doc_meta = config["DOC_METADATA"]
    pages_all = []

    pdf_files = sorted(sbp_dir.glob("*.pdf"))
    if not pdf_files:
        raise FileNotFoundError(
            f"No PDFs found in {sbp_dir}. Place SBP documents there first."
        )

    print(f"Found {len(pdf_files)} PDF files\n")


    for pdf_path in pdf_files:
        doc_name = pdf_path.name
        meta = doc_meta.get(doc_name, {"category": "SBP Regulation", "short_name": doc_name})

        try:
            reader    = PdfReader(str(pdf_path))
            doc_pages = []
            for i, page in enumerate(reader.pages):
                raw   = page.extract_text() or ""
                clean = clean_pdf_text(raw)
                if len(clean) < config["MIN_CHUNK_LENGTH"]:
                    continue
                doc_pages.append(SBPPage(
                    doc_name    = doc_name,
                    page_number = i + 1,
                    text        = clean,
                    category    = meta["category"],
                    short_name  = meta["short_name"],
                    section     = detect_section(clean),
                ))
            pages_all.extend(doc_pages)
            print(f"  {doc_name:<50} {len(reader.pages):>3} pages | {len(doc_pages):>3} usable | {meta['category']}")
        except Exception as e:
            print(f"  ERROR loading {doc_name}: {e}")

    print(f"\nTotal usable pages: {len(pages_all)}")
    return pages_all
````

## Section 5 — Chunking Strategy

Sliding-window character chunking with sentence-boundary snapping.  
`CHUNK_SIZE=500` chars, `CHUNK_OVERLAP=75` chars — configurable in Section 3.

````python
@dataclass
class Chunk:
    chunk_id    : str
    doc_name    : str
    page_number : int
    section     : str
    category    : str
    short_name  : str
    text        : str
    chunk_index : int
    char_start  : int
    char_end    : int

    def citation_label(self) -> str:
        return f"[{self.short_name}, {self.section}, Page {self.page_number}]"

    def source_entry(self) -> str:
        return (
            f"• {self.short_name} | {self.category} | "
            f"Page {self.page_number} | Section: {self.section}"
        )
````

````python
def chunk_text(text: str, chunk_size: int, chunk_overlap: int) -> List[Tuple[str, int, int]]:
    chunks = []
    start  = 0
    n      = len(text)
    while start < n:
        end = min(start + chunk_size, n)
        if end < n:
            zone = text[max(start, end - 50): end]
            last = max(zone.rfind('.'), zone.rfind('\n'), zone.rfind(';'))
            if last > 0:
                end = max(start, end - 50) + last + 1
        chunk = text[start:end].strip()
        if chunk:
            chunks.append((chunk, start, end))
        nxt = end - chunk_overlap
        start = nxt if nxt > start else start + 1
    return chunks
````

````python
def build_chunks(pages: List[SBPPage], config: dict) -> List[Chunk]:
    all_chunks = []
    for page in tqdm(pages, desc="Chunking pages"):
        windows = chunk_text(page.text, config["CHUNK_SIZE"], config["CHUNK_OVERLAP"])
        for idx, (text, start, end) in enumerate(windows):
            if len(text) < config["MIN_CHUNK_LENGTH"]:
                continue
            uid = hashlib.sha256(
                f"{page.doc_name}_{page.page_number}_{idx}_{text[:50]}".encode()
            ).hexdigest()[:16]
            all_chunks.append(Chunk(
                chunk_id    = uid,
                doc_name    = page.doc_name,
                page_number = page.page_number,
                section     = page.section,
                category    = page.category,
                short_name  = page.short_name,
                text        = text,
                chunk_index = idx,
                char_start  = start,
                char_end    = end,
            ))
    return all_chunks
````

## Section 6 — Embeddings Generation

````python
print(f"Loading embedding model: {CONFIG['EMBEDDING_MODEL']}...")
embed_model = SentenceTransformer(CONFIG["EMBEDDING_MODEL"])
print(f"Model loaded. Embedding dimension: {embed_model.get_sentence_embedding_dimension()}")
````

````python
def generate_embeddings(chunks: List[Chunk], model: SentenceTransformer, batch_size: int) -> np.ndarray:
    texts = [c.text for c in chunks]
    embs  = []
    for i in tqdm(range(0, len(texts), batch_size), desc="Embedding"):
        batch = texts[i: i + batch_size]
        embs.append(model.encode(
            batch,
            convert_to_numpy     = True,
            normalize_embeddings = True,
            show_progress_bar    = False,
        ))
    return np.vstack(embs)

t0               = time.time()
chunk_embeddings = generate_embeddings(all_chunks, embed_model, CONFIG["EMBEDDING_BATCH"])
elapsed          = time.time() - t0

print(f"Embeddings shape : {chunk_embeddings.shape}")
print(f"Time taken       : {elapsed:.1f}s  ({len(all_chunks)/elapsed:.0f} chunks/sec)")
````

## Section 7 — ChromaDB Vector Store

Persistent local store at `chroma_db/`. Re-running is safe — skips re-indexing if chunk count matches.

````python
def build_chroma_store(chunks: List[Chunk], embeddings: np.ndarray, config: dict) -> chromadb.Collection:
    client     = chromadb.PersistentClient(path=config["CHROMA_DB_PATH"])
    collection = client.get_or_create_collection(
        name     = config["COLLECTION_NAME"],
        metadata = {"hnsw:space": "cosine"},
    )
    existing = collection.count()
    print(f"Collection '{config['COLLECTION_NAME']}' — existing chunks: {existing}")

    if existing == len(chunks):
        print("Already fully indexed. Skipping.")
        return collection

    if existing > 0:
        print(f"Count mismatch ({existing} vs {len(chunks)}). Re-indexing...")
        client.delete_collection(config["COLLECTION_NAME"])
        collection = client.get_or_create_collection(
            name=config["COLLECTION_NAME"], metadata={"hnsw:space": "cosine"}
        )

    BATCH = 512
    print(f"Indexing {len(chunks)} chunks in batches of {BATCH}...")
    for i in tqdm(range(0, len(chunks), BATCH), desc="Indexing to ChromaDB"):
        bc = chunks[i: i + BATCH]
        be = embeddings[i: i + BATCH]
        collection.add(
            ids        = [c.chunk_id for c in bc],
            embeddings = be.tolist(),
            documents  = [c.text for c in bc],
            metadatas  = [{
                "doc_name"   : c.doc_name,
                "page_number": c.page_number,
                "section"    : c.section,
                "category"   : c.category,
                "short_name" : c.short_name,
                "chunk_index": c.chunk_index,
                "citation"   : c.citation_label(),
            } for c in bc],
        )

    print(f"Indexing complete. Total in store: {collection.count()}")
    return collection

chroma_collection = build_chroma_store(all_chunks, chunk_embeddings, CONFIG)
print(f"ChromaDB ready at: {CONFIG['CHROMA_DB_PATH']}")
````

## Section 8 — BM25 Keyword Index

In-memory BM25 over all chunks. Catches exact regulation numbers (e.g. Regulation-7, STR, CTR) that semantic search may miss.

````python
def tokenize_bm25(text: str) -> List[str]:
    return re.findall(r'\b[a-z0-9/\-]+\b', text.lower()) or ["<empty>"]

def build_bm25_index(chunks: List[Chunk]) -> BM25Okapi:
    tokenized = [tokenize_bm25(c.text) for c in chunks]
    return BM25Okapi(tokenized)

bm25_index = build_bm25_index(all_chunks)
print(f"BM25 index built over {len(all_chunks)} chunks.")

def bm25_retrieve(query: str, bm25: BM25Okapi, chunks: List[Chunk], top_k: int) -> List[Tuple[Chunk, float]]:
    scores     = bm25.get_scores(tokenize_bm25(query))
    top_idx    = np.argsort(scores)[::-1][:top_k]
    return [(chunks[i], float(scores[i])) for i in top_idx if scores[i] > 0]
````

## Section 9 — Hybrid Retrieval (Semantic + BM25 via RRF)

Reciprocal Rank Fusion merges both ranked lists without requiring score calibration between the two systems.

````python
def semantic_retrieve(
    query: str,
    collection: chromadb.Collection,
    model: SentenceTransformer,
    top_k: int,
    chunks_map: Dict[str, Chunk],
    categories: List[str] = None,
) -> List[Tuple[Chunk, float]]:
    
    qe = model.encode([query], convert_to_numpy=True, normalize_embeddings=True).tolist()
    
    query_kwargs = dict(
        query_embeddings = qe,
        n_results        = top_k,
        include          = ["documents", "metadatas", "distances"],
    )
    if categories:
        query_kwargs["where"] = {"category": {"$in": categories}}
    
    res = collection.query(**query_kwargs)
    
    results = []
    for doc_text, meta, dist in zip(res["documents"][0], res["metadatas"][0], res["distances"][0]):
        matched = None
        for c in chunks_map.values():
            if (
                c.doc_name    == meta["doc_name"]
                and c.page_number == meta["page_number"]
                and doc_text[:50] in c.text
            ):
                matched = c
                break
        if matched is None:
            matched = Chunk(
                chunk_id    = meta["doc_name"] + str(meta["page_number"]),
                doc_name    = meta["doc_name"],
                page_number = meta["page_number"],
                section     = meta["section"],
                category    = meta["category"],
                short_name  = meta["short_name"],
                text        = doc_text,
                chunk_index = meta.get("chunk_index", 0),
                char_start  = 0,
                char_end    = len(doc_text),
            )
        results.append((matched, 1.0 - dist))
    
    return results
````

````python
def reciprocal_rank_fusion(
    sem: List[Tuple[Chunk, float]],
    bm25: List[Tuple[Chunk, float]],
    k: int = 60,
) -> List[Tuple[Chunk, float]]:
    scores: Dict[str, float] = {}
    cmap:   Dict[str, Chunk] = {}
    for rank, (chunk, _) in enumerate(sem):
        scores[chunk.chunk_id] = scores.get(chunk.chunk_id, 0.0) + 1.0 / (k + rank + 1)
        cmap[chunk.chunk_id]   = chunk
    for rank, (chunk, _) in enumerate(bm25):
        scores[chunk.chunk_id] = scores.get(chunk.chunk_id, 0.0) + 1.0 / (k + rank + 1)
        cmap[chunk.chunk_id]   = chunk
    return [(cmap[uid], scores[uid]) for uid in sorted(scores, key=lambda x: scores[x], reverse=True)]
````

````python
def hybrid_retrieve(
    query: str, collection: chromadb.Collection,
    model: SentenceTransformer, bm25: BM25Okapi,
    chunks: List[Chunk], config: dict,
) -> List[Tuple[Chunk, float]]:
    chunks_map    = {c.chunk_id: c for c in chunks}
    sem_results   = semantic_retrieve(query, collection, model, config["TOP_K_SEMANTIC"], chunks_map, categories = ["AML/CFT/CPF", "Branchless Banking"])
    bm25_results  = bm25_retrieve(query, bm25, chunks, config["TOP_K_BM25"])
    return reciprocal_rank_fusion(sem_results, bm25_results)
````

## Section 10 — Reranking & Context Filtering

Cross-encoder scores each (query, chunk) pair for genuine relevance. Falls back to RRF scores if unavailable.

````python
class Reranker:
    MODEL = "cross-encoder/ms-marco-MiniLM-L-6-v2"

    def __init__(self):
        self.model     = None
        self.available = False
        try:
            print(f"Loading cross-encoder: {self.MODEL}...")
            self.model     = CrossEncoder(self.MODEL)
            self.available = True
            print("Cross-encoder loaded.")
        except Exception as e:
            print(f"Cross-encoder unavailable ({e}). Falling back to RRF scores.")

    def rerank(self, query: str, results: List[Tuple[Chunk, float]], top_k: int) -> List[Tuple[Chunk, float]]:
        if not results:
            return []
        if self.available:
            pairs  = [(query, c.text) for c, _ in results]
            scores = self.model.predict(pairs).tolist()
            return sorted(zip([r[0] for r in results], scores), key=lambda x: x[1], reverse=True)[:top_k]
        return results[:top_k]


def filter_context(results: List[Tuple[Chunk, float]]) -> List[Tuple[Chunk, float]]:
    seen, out = set(), []
    for chunk, score in results:
        key = (chunk.doc_name, chunk.page_number)
        if key not in seen:
            seen.add(key)
            out.append((chunk, score))
    return out

reranker = Reranker()
````

## Section 11 — Query Builder & Query Expansion

Converts XGBoost output (fraud_probability + features) into a natural language query. Expansion appends SBP-specific regulatory terms to improve BM25 recall.

````python
FEATURE_NAMES = [
    "step", "amount", "oldbalanceOrg", "newbalanceOrig",
    "oldbalanceDest", "newbalanceDest", "balanceDiff", "amount_ratio",
    "type_CASH_OUT", "type_DEBIT", "type_PAYMENT", "type_TRANSFER",
]

RISK_TIERS = {
    "CRITICAL": (0.85, 1.01),
    "HIGH"    : (0.65, 0.85),
    "MEDIUM"  : (0.50, 0.65),
}
````

````python
def get_risk_tier(prob: float) -> str:
    for tier, (lo, hi) in RISK_TIERS.items():
        if lo <= prob < hi:
            return tier
    return "LOW"
````

````python
def decode_tx_type(features: Dict[str, float]) -> str:
    for col, name in [("type_CASH_OUT","CASH_OUT"),("type_DEBIT","DEBIT"),
                      ("type_PAYMENT","PAYMENT"),("type_TRANSFER","TRANSFER")]:
        if features.get(col, 0) >= 0.5:
            return name
    return "CASH_IN"
````

````python
def build_transaction_query(prob: float, features: Dict[str, float], expand: bool = True) -> str:
    tx_type   = decode_tx_type(features)
    amount    = features.get("amount", 0)
    bal_diff  = features.get("balanceDiff", 0)
    amt_ratio = features.get("amount_ratio", 0)
    old_bal   = features.get("oldbalanceOrg", 0)
    new_bal   = features.get("newbalanceOrig", 0)
    risk      = get_risk_tier(prob)

    base = (
        f"A {tx_type} transaction of PKR {amount:,.0f} has been flagged as potentially fraudulent "
        f"with a {risk} risk score of {prob:.1%}. "
        f"The origin account balance changed from PKR {old_bal:,.0f} to PKR {new_bal:,.0f} "
        f"(balance difference: PKR {bal_diff:,.0f}, amount-to-balance ratio: {amt_ratio:.2f}). "
        f"What SBP regulations apply and what reporting obligations exist?"
    )

    if not expand:
        return base

    expansion = []
    if tx_type in ("CASH_OUT", "TRANSFER"):
        expansion += ["suspicious transaction report STR AML CFT money laundering",
                      "customer due diligence CDD enhanced due diligence EDD wire transfer fund transfer"]
    if amount > 1_000_000:
        expansion += ["currency transaction report CTR large transaction threshold unusual large"]
    if amt_ratio > 0.9 or (old_bal > 0 and new_bal == 0):
        expansion += ["account draining high risk risk based approach internal risk assessment IRAR"]
    if risk == "CRITICAL":
        expansion += ["immediate STR reporting Regulation-7 suspicious transaction regulated entity SBP compliance"]

    return (base + " " + " ".join(expansion)).strip()
````

## Section 12 — Grounded Prompt Templates

Designed to force citation, prevent hallucination, and produce structured output (inline citations + sources list).

````python
SYSTEM_PROMPT = """You are a regulatory compliance assistant specializing in State Bank of Pakistan (SBP) regulations.

Your role is to analyze flagged financial transactions and provide legally grounded justifications based ONLY on the provided SBP regulatory context.

STRICT RULES:
1. Ground every claim in the provided SBP context. Do NOT use general knowledge.
2. Cite sources inline as [Document Name, Section, Page X].
3. If context is insufficient, state explicitly: "No grounded SBP regulatory evidence found."
4. Do NOT hallucinate regulation numbers, section names, or page references.
5. Responses must be factual, concise, and professional English.
6. End every response with a SOURCES section listing all cited documents.
"""

REGULATORY_PROMPT_TEMPLATE = """
## TRANSACTION RISK ASSESSMENT

**Transaction Details:**
{transaction_summary}

**XGBoost Fraud Probability:** {fraud_probability:.1%}
**Risk Tier:** {risk_tier}

---

## RETRIEVED SBP REGULATORY CONTEXT

{regulatory_context}

---

## YOUR TASK

Based SOLELY on the SBP context above, provide:

1. **Regulatory Assessment** — Which SBP regulations are triggered? Cite each as [Document, Section, Page].
2. **Reporting Obligations** — STR/CTR requirements and deadlines if stated in context.
3. **Compliance Actions Required** — Immediate actions the financial institution must take.
4. **Risk Justification** — Why does this transaction pattern raise regulatory concern?
5. **Regulatory Basis Summary** — Key SBP regulatory grounds for flagging this transaction.

End with:
**SOURCES:**
• Document Name | Category | Page | Section
(list every cited source)

If context is insufficient for any point, state: "No grounded SBP regulatory evidence found."
"""
````

````python
def format_regulatory_context(chunks_with_scores: List[Tuple[Chunk, float]]) -> str:
    blocks = []
    for i, (chunk, score) in enumerate(chunks_with_scores, 1):
        blocks.append(f"[Context {i}] {chunk.citation_label()}\n{chunk.text}")
    return "\n\n".join(blocks)
````

````python
def format_tx_summary(prob: float, features: Dict[str, float]) -> str:
    tx_type   = decode_tx_type(features)
    amount    = features.get("amount", 0)
    bal_diff  = features.get("balanceDiff", 0)
    amt_ratio = features.get("amount_ratio", 0)
    old_bal   = features.get("oldbalanceOrg", 0)
    new_bal   = features.get("newbalanceOrig", 0)
    return (
        f"- Transaction Type     : {tx_type}\n"
        f"- Amount               : PKR {amount:,.0f}\n"
        f"- Origin Balance Before: PKR {old_bal:,.0f}\n"
        f"- Origin Balance After : PKR {new_bal:,.0f}\n"
        f"- Balance Difference   : PKR {bal_diff:,.0f}\n"
        f"- Amount/Balance Ratio : {amt_ratio:.2f}\n"
        f"- Risk Tier            : {get_risk_tier(prob)}"
    )
````

## Section 13 — Grounded Response Generation

````python
@dataclass
class RAGResponse:
    transaction_id    : str
    fraud_probability : float
    risk_tier         : str
    query             : str
    retrieved_chunks  : List[Chunk]
    llm_response      : str
    inline_citations  : List[str]
    sources           : List[str]
    latency_seconds   : float
    grounding_score   : float
    no_evidence_flag  : bool
````

````python
def extract_citations(text: str) -> List[str]:
    pattern = r'\[([^\[\]]+,\s*[^\[\]]+,\s*Page\s*\d+[^\[\]]*)\]'
    return list(dict.fromkeys(re.findall(pattern, text)))
````

````python
def compute_grounding_score(text: str) -> float:
    sentences = [s.strip() for s in re.split(r'[.!?]', text) if len(s.strip()) > 20]
    if not sentences:
        return 0.0
    cited = sum(1 for s in sentences if '[' in s and ']' in s and 'Page' in s)
    return cited / len(sentences)
````

````python
def check_hallucination(text: str, chunks: List[Chunk]) -> bool:
    cited_pages     = set(re.findall(r'Page\s*(\d+)', text))
    available_pages = set(str(c.page_number) for c in chunks)
    return bool(cited_pages - available_pages)
````

````python
from urllib import response


def generate_regulatory_response(
    fraud_probability : float,
    features          : Dict[str, float],
    retrieved_chunks  : List[Tuple[Chunk, float]],
    client            : openai.OpenAI,
    config            : dict,
    transaction_id    : str = "TXN-001",
) -> RAGResponse:
    t_start    = time.time()
    risk_tier  = get_risk_tier(fraud_probability)
    chunks_only = [c for c, _ in retrieved_chunks]

    reg_context = format_regulatory_context(retrieved_chunks)
    tx_summary  = format_tx_summary(fraud_probability, features)

    user_prompt = REGULATORY_PROMPT_TEMPLATE.format(
        transaction_summary = tx_summary,
        fraud_probability   = fraud_probability,
        risk_tier           = risk_tier,
        regulatory_context  = reg_context,
    )

    try:
        completion = client.chat.completions.create(
            model       = config["OPENAI_MODEL"],
            messages    = [
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user",   "content": user_prompt},
            ],
            max_tokens  = config["MAX_TOKENS"],
            temperature = config["TEMPERATURE"],
        )
        llm_text = completion.choices[0].message.content.strip()
    except openai.AuthenticationError:
        llm_text = "ERROR: Invalid OpenAI API key. Set OPENAI_API_KEY environment variable."
    except openai.RateLimitError:
        llm_text = "ERROR: OpenAI rate limit reached. Wait and retry."
    except Exception as e:
        llm_text = f"ERROR: LLM call failed — {e}"

    latency     = time.time() - t_start
    citations   = extract_citations(llm_text)
    grounding   = compute_grounding_score(llm_text)
    no_evidence = "no grounded" in llm_text.lower() or "insufficient" in llm_text.lower()
    sources     = list(dict.fromkeys(c.source_entry() for c in chunks_only))

    return RAGResponse(
        transaction_id    = transaction_id,
        fraud_probability = fraud_probability,
        risk_tier         = risk_tier,
        query             = build_transaction_query(fraud_probability, features),
        retrieved_chunks  = chunks_only,
        llm_response      = llm_text,
        inline_citations  = citations,
        sources           = sources,
        latency_seconds   = latency,
        grounding_score   = grounding,
        no_evidence_flag  = no_evidence,
    )
````

## Section 14 — Full RAG Pipeline

````python
def run_rag_pipeline(
    fraud_probability : float,
    features          : Dict[str, float],
    collection        : chromadb.Collection,
    embed_model       : SentenceTransformer,
    bm25              : BM25Okapi,
    chunks            : List[Chunk],
    reranker          : Reranker,
    client            : openai.OpenAI,
    config            : dict,
    transaction_id    : str = "TXN-001",
    verbose           : bool = True,
) -> Optional['RAGResponse']:

    if fraud_probability < config["FRAUD_THRESHOLD"]:
        if verbose:
            print(f"[{transaction_id}] Prob {fraud_probability:.3f} below threshold. Skipping.")
        return None

    if verbose:
        print(f"\n{'='*65}")
        print(f"  RAG: {transaction_id}  |  Prob={fraud_probability:.1%}  |  Risk={get_risk_tier(fraud_probability)}  |  Type={decode_tx_type(features)}")
        print(f"{'='*65}")

    # 1. Query
    query = build_transaction_query(fraud_probability, features, expand=True)
    if verbose: print(f"[1/5] Query built ({len(query)} chars)")

    # 2. Hybrid retrieval
    hybrid = hybrid_retrieve(query, collection, embed_model, bm25, chunks, config)
    if verbose: print(f"[2/5] Hybrid retrieval: {len(hybrid)} candidates")

    rerank_query = (
        f"SBP regulations for suspicious {decode_tx_type(features)} transaction "
        f"fraud probability {fraud_probability:.0%}. STR reporting obligations AML CFT."    
    )

    # 3. Rerank
    reranked = reranker.rerank(rerank_query, hybrid, top_k=config["TOP_K_FINAL"] + 5)
    if verbose: print(f"[3/5] Reranked: {len(reranked)} results")

    # 4. Filter
    filtered = filter_context(reranked)[:config["TOP_K_FINAL"]]
    if verbose:
        print(f"[4/5] Final context: {len(filtered)} chunks")
        for i, (c, s) in enumerate(filtered, 1):
            print(f"      {i}. {c.citation_label()} (score={s:.3f})")

    # 5. Generate
    if verbose: print(f"[5/5] Generating via {config['OPENAI_MODEL']}...")
    response = generate_regulatory_response(
        fraud_probability=fraud_probability, features=features,
        retrieved_chunks=filtered, client=client, config=config,
        transaction_id=transaction_id,
    )

    if verbose:
        print(f"  Done in {response.latency_seconds:.1f}s | Grounding={response.grounding_score:.2f} | Citations={len(response.inline_citations)}")

    return response
````

````python
def display_rag_response(r: RAGResponse):
    print(f"\n{'='*70}")
    print(f"  REGULATORY JUSTIFICATION — {r.transaction_id}")
    print(f"{'='*70}")
    print(f"  Fraud Prob      : {r.fraud_probability:.1%}")
    print(f"  Risk Tier       : {r.risk_tier}")
    print(f"  Grounding Score : {r.grounding_score:.2f}")
    print(f"  Citations Found : {len(r.inline_citations)}")
    print(f"  Latency         : {r.latency_seconds:.1f}s")
    print(f"  No-Evidence Flag: {r.no_evidence_flag}")
    print(f"{'─'*70}")
    print(r.llm_response)
    print(f"{'─'*70}")
    print("RETRIEVED SOURCES:")
    for src in r.sources:
        print(f"  {src}")
    print(f"{'='*70}")
````

````python
def evaluate_retrieval(eval_queries, collection, embed_model, bm25, chunks, reranker, config):
    rows = []
    chunks_map = {c.chunk_id: c for c in chunks}
    for i, eq in enumerate(eval_queries, 1):
        q      = eq["query"]
        e_docs = set(eq["expected_docs"])
        e_terms= [t.lower() for t in eq["expected_terms"]]

        hybrid   = hybrid_retrieve(q, collection, embed_model, bm25, chunks, config)
        reranked = reranker.rerank(q, hybrid, config["TOP_K_FINAL"])
        filtered = filter_context(reranked)

        ret_docs = [c.doc_name for c, _ in filtered]
        ret_text = " ".join(c.text.lower() for c, _ in filtered)

        prec      = sum(1 for d in ret_docs if d in e_docs) / max(len(filtered), 1)
        term_rate = sum(1 for t in e_terms if t in ret_text) / max(len(e_terms), 1)
        has_doc   = any(d in e_docs for d in ret_docs)

        rows.append({
            "q_id"            : i,
            "query"           : q[:60]+"...",
            "precision_at_5"  : round(prec, 3),
            "term_hit_rate"   : round(term_rate, 3),
            "has_expected_doc": has_doc,
            "n_retrieved"     : len(filtered),
        })
    return pd.DataFrame(rows)
````

````python
def evaluate_responses(responses):
    rows = []
    for r in responses:
        rows.append({
            "transaction_id" : r.transaction_id,
            "fraud_prob"     : round(r.fraud_probability, 3),
            "risk_tier"      : r.risk_tier,
            "grounding_score": round(r.grounding_score, 3),
            "citation_count" : len(r.inline_citations),
            "sources_count"  : len(r.sources),
            "hallucination"  : check_hallucination(r.llm_response, r.retrieved_chunks),
            "no_evidence"    : r.no_evidence_flag,
            "latency_s"      : round(r.latency_seconds, 2),
        })
    return pd.DataFrame(rows)
````

## Section 19 — Batch Processing from Pipeline Outputs

````python
def predict_and_rag(
    transactions_df  : pd.DataFrame,
    xgb_model,
    scaler,
    collection, embed_model, bm25, chunks, reranker, client, config,
    feature_names    : List[str] = None,
    max_transactions : int = 10,
) -> List[RAGResponse]:

    if feature_names is None:
        feature_names = FEATURE_NAMES

    if xgb_model is None:
        print("No XGBoost model. Using synthetic transactions.")
        responses = []
        for tx in SYNTHETIC_TRANSACTIONS[:max_transactions]:
            r = run_rag_pipeline(
                tx["fraud_prob"], tx["features"], collection, embed_model,
                bm25, chunks, reranker, client, config, tx["id"], verbose=False,
            )
            if r: responses.append(r)
        return responses

    avail = [f for f in feature_names if f in transactions_df.columns]
    X     = transactions_df[avail].values.astype("float32")
    if scaler: X = scaler.transform(X)

    probs       = xgb_model.predict_proba(X)[:, 1]
    flagged_idx = np.where(probs >= config["FRAUD_THRESHOLD"])[0]
    print(f"Transactions: {len(transactions_df)} | Flagged: {len(flagged_idx)} | Processing: {min(len(flagged_idx), max_transactions)}")

    responses = []
    for i, idx in enumerate(flagged_idx[:max_transactions]):
        row      = transactions_df.iloc[idx]
        features = {f: float(row[f]) for f in avail}
        prob     = float(probs[idx])
        tx_id    = f"TXN-{idx:06d}"
        r = run_rag_pipeline(prob, features, collection, embed_model, bm25, chunks,
                             reranker, client, config, tx_id, verbose=False)
        if r:
            responses.append(r)
            print(f"  [{i+1}] {tx_id} | Prob={prob:.1%} | Citations={len(r.inline_citations)} | {r.latency_seconds:.1f}s")
    return responses

# Demo with synthetic data
print("Batch demo (2 synthetic transactions):")
batch = predict_and_rag(
    pd.DataFrame(), None, scaler,
    chroma_collection, embed_model, bm25_index, all_chunks, reranker,
    openai_client, CONFIG, max_transactions=2,
)
print(f"Batch complete. {len(batch)} responses.")
````

````python
# At the top of rag_module.py, after imports:
try:
    from dotenv import load_dotenv
    load_dotenv(dotenv_path=r"C:\Users\pakistan\OneDrive\Desktop\ai_project\.env.txt")
except ImportError:
    pass

module_path = Path(CONFIG["PROJECT_ROOT"]) / "rag_module.py"

module_code = """
import os, re, time
import numpy as np
from dataclasses import dataclass
from typing import List, Dict, Optional
import chromadb
from sentence_transformers import SentenceTransformer, CrossEncoder
from rank_bm25 import BM25Okapi
import openai

RAG_CONFIG = {
    "CHROMA_DB_PATH"  : r"C:\\Users\\pakistan\\OneDrive\\Desktop\\ai_project\\chroma_db",
    "COLLECTION_NAME" : "sbp_regulations",
    "EMBEDDING_MODEL" : "sentence-transformers/all-MiniLM-L6-v2",
    "OPENAI_MODEL"    : "gpt-4o-mini",
    "MAX_TOKENS"      : 1000,
    "TEMPERATURE"     : 0.1,
    "TOP_K_SEMANTIC"  : 10,
    "TOP_K_BM25"      : 10,
    "TOP_K_FINAL"     : 5,
    "FRAUD_THRESHOLD" : 0.5,
}

RISK_TIERS = {"CRITICAL": (0.85, 1.01), "HIGH": (0.65, 0.85), "MEDIUM": (0.50, 0.65)}

def get_risk_tier(prob):
    for tier, (lo, hi) in RISK_TIERS.items():
        if lo <= prob < hi:
            return tier
    return "LOW"

def decode_tx_type(features):
    for col, name in [("type_CASH_OUT","CASH_OUT"), ("type_DEBIT","DEBIT"),
                      ("type_PAYMENT","PAYMENT"),   ("type_TRANSFER","TRANSFER")]:
        if features.get(col, 0) >= 0.5:
            return name
    return "CASH_IN"

def _tokenize(text):
    return re.findall(r"\b[a-z0-9/\-]+\b", text.lower()) or ["<empty>"]

@dataclass
class StreamlitRAGResult:
    transaction_id   : str
    fraud_probability: float
    risk_tier        : str
    response_text    : str
    sources          : list
    citations        : list
    grounding_score  : float
    latency_seconds  : float
    no_evidence_flag : bool
    retrieved_chunks : list

_components = {}

def load_rag_components(openai_api_key=None):
    global _components
    if _components:
        return _components
    if openai_api_key:
        os.environ["OPENAI_API_KEY"] = openai_api_key
    embed_model   = SentenceTransformer(RAG_CONFIG["EMBEDDING_MODEL"])
    chroma_client = chromadb.PersistentClient(path=RAG_CONFIG["CHROMA_DB_PATH"])
    collection    = chroma_client.get_collection(RAG_CONFIG["COLLECTION_NAME"])
    client        = openai.OpenAI(api_key=os.environ.get("OPENAI_API_KEY", ""))
    all_data      = collection.get(include=["documents", "metadatas"])
    texts         = all_data["documents"]
    metas         = all_data["metadatas"]
    bm25          = BM25Okapi([_tokenize(t) for t in texts])
    try:
        reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
    except Exception:
        reranker = None
    _components = {
        "embed_model": embed_model, "collection": collection,
        "client": client, "bm25": bm25, "texts": texts,
        "metas": metas, "reranker": reranker,
    }
    return _components

def rag_pipeline_for_streamlit(fraud_probability, features, transaction_id="TXN-STREAMLIT", openai_api_key=None):
    comp        = load_rag_components(openai_api_key)
    embed_model = comp["embed_model"]
    collection  = comp["collection"]
    client      = comp["client"]
    bm25        = comp["bm25"]
    texts       = comp["texts"]
    metas       = comp["metas"]
    reranker    = comp["reranker"]

    tx_type   = decode_tx_type(features)
    risk_tier = get_risk_tier(fraud_probability)
    amount    = features.get("amount", 0)
    old_bal   = features.get("oldbalanceOrg", 0)
    new_bal   = features.get("newbalanceOrig", 0)

    query = (
        f"A {tx_type} transaction of PKR {amount:,.0f} flagged "
        f"({fraud_probability:.1%} {risk_tier} risk). "
        f"Balance PKR {old_bal:,.0f} to PKR {new_bal:,.0f}. "
        "SBP AML CFT regulations STR reporting suspicious transaction CDD EDD."
    )

    t0  = time.time()
    qe  = embed_model.encode([query], convert_to_numpy=True, normalize_embeddings=True).tolist()
    fraud_categories = ["AML/CFT/CPF", "Branchless Banking"]
    sem = collection.query(
        query_embeddings=qe,
        n_results=RAG_CONFIG["TOP_K_SEMANTIC"],
        where={"category": {"$in": fraud_categories}},
        include=["documents", "metadatas", "distances"],
    )
    scores   = bm25.get_scores(_tokenize(query))
    bm25_top = np.argsort(scores)[::-1][:RAG_CONFIG["TOP_K_BM25"]]

    cands = {}
    for rank, (doc, meta, dist) in enumerate(zip(
        sem["documents"][0], sem["metadatas"][0], sem["distances"][0]
    )):
        key = (meta["doc_name"], meta["page_number"])
        cands[key] = {"text": doc, "meta": meta, "score": 1.0 / (61 + rank)}

    for rank, idx in enumerate(bm25_top):
        if scores[idx] > 0:
            meta = metas[idx]
            key  = (meta["doc_name"], meta["page_number"])
            if key not in cands:
                cands[key] = {"text": texts[idx], "meta": meta, "score": 0.0}
            cands[key]["score"] += 1.0 / (61 + rank)

    sorted_c = sorted(cands.values(), key=lambda x: x["score"], reverse=True)

    if reranker and sorted_c:
        pairs    = [(query, c["text"]) for c in sorted_c]
        rs       = reranker.predict(pairs).tolist()
        sorted_c = [c for _, c in sorted(zip(rs, sorted_c), key=lambda x: x[0], reverse=True)]

    final          = sorted_c[:RAG_CONFIG["TOP_K_FINAL"]]
    context_blocks = []
    sources        = []
    chunks_display = []

    for i, c in enumerate(final, 1):
        m    = c["meta"]
        cite = m.get("citation", f"[{m.get('short_name','SBP')}, Page {m.get('page_number',0)}]")
        context_blocks.append(f"[Context {i}] {cite}\\n{c['text']}")
        sources.append(
            f"- {m.get('short_name')} | {m.get('category')} | "
            f"Page {m.get('page_number')} | {m.get('section')}"
        )
        chunks_display.append({
            "citation": cite,
            "text"    : c["text"][:300] + "...",
            "page"    : m.get("page_number"),
        })

    context_str = "\\n\\n".join(context_blocks)
    system = (
        "You are an SBP regulatory compliance assistant. "
        "Ground every claim in the provided context. "
        "Cite sources as [Document, Section, Page X]. Never hallucinate. "
        "If context is insufficient, state: No grounded SBP regulatory evidence found."
    )
    user = (
        f"Transaction: {tx_type} | PKR {amount:,.0f} | {fraud_probability:.1%} | {risk_tier}\\n\\n"
        f"RETRIEVED SBP CONTEXT:\\n{context_str}\\n\\n"
        "Provide: 1) Regulations triggered (cite each), "
        "2) Reporting obligations (STR/CTR), "
        "3) Compliance actions, 4) Risk justification, "
        "5) Regulatory basis summary. End with SOURCES."
    )

    try:
        completion    = client.chat.completions.create(
            model     = RAG_CONFIG["OPENAI_MODEL"],
            messages  = [
                {"role": "system", "content": system},
                {"role": "user",   "content": user},
            ],
            max_tokens  = RAG_CONFIG["MAX_TOKENS"],
            temperature = RAG_CONFIG["TEMPERATURE"],
        )
        response_text = completion.choices[0].message.content.strip()
    except Exception as e:
        response_text = f"ERROR: {e}"

    latency   = time.time() - t0
    pattern   = r"\\[([^\\[\\]]+,\\s*[^\\[\\]]+,\\s*Page\\s*\\d+[^\\[\\]]*)\\]"
    citations = list(dict.fromkeys(re.findall(pattern, response_text)))
    sents     = [s.strip() for s in re.split(r"[.!?]", response_text) if len(s.strip()) > 20]
    grounding = sum(1 for s in sents if "[" in s and "]" in s) / max(len(sents), 1)
    no_ev     = "no grounded" in response_text.lower()

    return StreamlitRAGResult(
        transaction_id    = transaction_id,
        fraud_probability = fraud_probability,
        risk_tier         = risk_tier,
        response_text     = response_text,
        sources           = list(dict.fromkeys(sources)),
        citations         = citations,
        grounding_score   = round(grounding, 3),
        latency_seconds   = round(latency, 2),
        no_evidence_flag  = no_ev,
        retrieved_chunks  = chunks_display,
    )
"""

with open(module_path, "w", encoding="utf-8") as f:
    f.write(module_code.strip())

print(f"Streamlit module saved: {module_path}")
print("Usage: from rag_module import load_rag_components, rag_pipeline_for_streamlit")
````
